```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
```

### Definiciones

Asíntota: Línea recta que, prolongada indefinidamente, se acerca progresivamente a una curva sin llegar nunca a encontrarla.

```python
fig, ax = plt.subplots(2, 2, sharex=True)
x = np.linspace(0, 10, 101)

def generator(n):
    def f(x):
        result = 1
        for i in range(1, n+1):
            result *= (x - i)
        return 5 / result

    return f

def f(x):
    return (5/(x-1))

def g(x):
    return (5/((x-1)*(x-2)))

g_max = ([i/10 for i in argrelextrema(g(x), np.greater)])

def h(x):
    return (5/((x-1)*(x-2)*(x-3)))

def p(x):
    return (5/((x-1)*(x-2)*(x-3)*(x-4)))


ax[0, 0].plot(x, f(x))
ax[0, 1].plot(x, g(x), marker="o", markevery=[x.tolist().index(i) for i in np.array(g_max).flatten()])
ax[1, 0].plot(x, h(x))
ax[1, 1].plot(x, p(x))
ax[0, 0].axvline([1], color='r')
ax[0, 1].axvline([1], color='r')
ax[0, 1].axvline([2], color='r')
ax[1, 0].axvline([1], color='r')
ax[1, 0].axvline([2], color='r')
ax[1, 0].axvline([3], color='r')
ax[1, 1].axvline([1], color='r')
ax[1, 1].axvline([2], color='r')
ax[1, 1].axvline([3], color='r')
ax[1, 1].axvline([4], color='r')
ax[0, 0].set_title(r'$f(x) = \frac{5}{(x-1)}$')
ax[0, 1].set_title(r'$g(x) = \frac{5}{(x-1)(x-2)}$')
ax[1, 0].set_title(r'$h(x) = \frac{5}{(x-1)(x-2)(x-3)}$')
ax[1, 1].set_title(r'$p(x) = \frac{5}{(x-1)(x-2)(x-3)(x-4)}$')
```

### Patrones evidentes

Podemos evidenciar que las asíntotas de las funciones existen a intervalos constantes de 1. 
Adicionalmente, para las funciones que terminan con un factor par, el mínimo o máximo de la función se encuentra en la mitad.
