```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
```

## Introducción

En esta exploración, voy a encontrar patrones y consistencias en las funciones racionales de la forma
$f(x) = \frac{5}{(x-1)...(x-r)}$
incluyendo asíntotas horizontales, verticales, y oblicuas.

### Definiciones

- Asíntota: Línea recta que, prolongada indefinidamente, se acerca progresivamente a una curva sin llegar nunca a encontrarla. Puede ser horizontal, vertical, o oblicua. Estos son líneas paralelas al eje x, y, o ninguno, respectivamente.
- Función racional: Funcion de la forma $R(x) = \frac{P(x)}{Q(x)}$ donde $Q(x) \neq 0$

```python
fig, ax = plt.subplots(2, 2, sharex=True)
x = np.linspace(0, 10, 101)

def generator(n):
    def f(x):
        result = 1
        for i in range(1, n+1):
            result *= (x - i) 
        return (5 / result)

    return f

g_max = ([i/10 for i in argrelextrema(generator(2)(x), np.greater)])

ax[0, 0].plot(x, generator(1)(x))
ax[0, 1].plot(x, generator(2)(x))
ax[1, 0].plot(x, generator(3)(x))
ax[1, 1].plot(x, generator(4)(x))
ax[0, 0].axvline([1], color='r')
ax[0, 1].axvline([1], color='r')
ax[0, 1].axvline([2], color='r')
ax[1, 0].axvline([1], color='r')
ax[1, 0].axvline([2], color='r')
ax[1, 0].axvline([3], color='r')
ax[1, 1].axvline([1], color='r')
ax[1, 1].axvline([2], color='r')
ax[1, 1].axvline([3], color='r')
ax[1, 1].axvline([4], color='r')
ax[0, 0].axhline([0], color="r")
ax[0, 1].axhline([0], color="r")
ax[1, 0].axhline([0], color="r")
ax[1, 1].axhline([0], color="r")
ax[0, 0].set_title(r'$f(x) = \frac{5}{(x-1)}$')
ax[0, 1].set_title(r'$g(x) = \frac{5}{(x-1)(x-2)}$')
ax[1, 0].set_title(r'$h(x) = \frac{5}{(x-1)(x-2)(x-3)}$')
ax[1, 1].set_title(r'$p(x) = \frac{5}{(x-1)(x-2)(x-3)(x-4)}$')
```

#### Asíntotas horizontales

En una función racional $R(x) = \frac{P(x)}{Q(x)}$, las asíntotas horizontales solo existen si el grado de $P(x)$ es menor o igual al grado de $Q(x)$. En las funciones que estamos analizando, puesto que el grado del numerador siempre es 0, al considerar a 5 como $(5)(x^0)$ y el grado del denominador siempre es por lo menos 1, podemos determinar que la asíntota horizontal de $R(x)$ siempre se encuentra en $y = 0$

#### Asíntotas verticales

En una función racional $R(x) = \frac{P(x)}{Q(x)}$, las asíntotas verticales existen en las raíces de $Q(x)$. Por lo tanto, dado a que $Q(x)$ siempre es una factorización de un polinomio en las funciones que estamos analizando, siempre van a existir tantas asíntotas verticales como hay factores lineales en el denominador, en cada solución del polinomio.

#### Asíntotas oblicuas

En una función racional $R(x) = \frac{P(x)}{Q(x)}$, las asíntotas oblicuas existen si el grado de $P(x)$ es mayor al grado de $Q(x)$. Como fue establecido anteriormente, en las funciones que estamos analizando, esto nunca ocurre, por lo que las funciones nunca tendrán asíntotas oblicuas.
